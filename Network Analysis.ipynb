```python
# getting text
file = open('Fictions/MoYan_Frog _CH1.txt')
text = file.read()
text
```

```python
import jieba
import jieba.analyse
nodes = jieba.analyse.extract_tags(text, topK=100, withWeight=False, allowPOS=('ns'))
print (len(nodes))
nodes
```

```python
import re
# tokenize into sentences
def split_sentences(full_text):
    sents = re.split(u'[\n。]', full_text)
    sents = [sent for sent in sents if len(sent) > 0]  # 去除只包含\n或空白符的句子
    return sents
```

```python
sents = split_sentences(text)
sents
```

```python
edges = []
nodes_new = []
for sent in sents:
    for node1 in nodes:
        for node2 in nodes:
            if node1 in sent:
                if node2 in sent:
                    if node1 != node2:
                        edges.append((node1,node2))
                        nodes_new.append(node1)
                        nodes_new.append(node2)
                continue
            continue
nodes_new = set(nodes_new)
edges
```

```python
import networkx as nx
import matplotlib.pyplot as plt
G = nx.karate_club_graph()
#G = nx.Graph()
G.add_nodes_from(nodes_new)
G.add_edges_from(edges)


nx.draw_circular(G, with_labels=True)
plt.show()

'''nx.draw(G,with_labels=True, node_size=40)  
plt.figure(figsize=(500,400))
plt.show()'''
```

```python
import networkx as nx  
import matplotlib.pyplot as plt  
  
#定义有向图  
DG = nx.DiGraph()   
#添加五个节点(列表)  
DG.add_nodes_from(['A', 'B', 'C', 'D', 'E'])  

#添加边(列表)  
DG.add_edges_from([('A', 'B'), ('A', 'C'), ('A', 'D'), ('D','A'),('E','A'),('E','D')])  

#绘制图形 设置节点名显示\节点大小\节点颜色  
colors = ['red', 'green', 'blue', 'yellow']  
nx.draw(DG,with_labels=True, node_size=900, node_color = colors)  
plt.show()
```

```python
elarge = [(u, v) for (u, v, d) in G.edges(data=True) if d['weight'] > 3]
esmall = [(u, v) for (u, v, d) in G.edges(data=True) if d['weight'] <= 3]
```

```python
import matplotlib
matplotlib.rcParams['font.sans-serif'] = 'SimHei'
pos = nx.spring_layout(G)  # positions for all nodes

# nodes
nx.draw_networkx_nodes(G, pos, node_size=700)

# edges
nx.draw_networkx_edges(G, pos, edgelist=elarge,
                       width=6)
nx.draw_networkx_edges(G, pos, edgelist=esmall,
                       width=6, alpha=0.5, edge_color='b', style='dashed')

# labels
nx.draw_networkx_labels(G, pos, font_size=8, font_family='sans-serif')

plt.axis('off')
plt.show()
```
