## Visualizing Topic Shares

This notebook is to explore topics in nine Chinese fictions written by three authors. Specifically, 7 topics are extracted from the nine fictions. We can have a general idea of the topics from the keywords extracted from each topic. In addition, a heatmap was generated to show the topic share the nine fictions. 

### Getting text
In this sample, texts are nine ebooks of three authors:
- JiaPingwa_happyDreams.txt
- JiaPingwa_oldFurnace.txt
- JiaPingwa_ruinedCity.txt
- MoYan_Frog.txt
- MoYan_bigBreast&wideHips.txt
- MoYan_redSorghum.txt
- WangShuo_dangerousAnimals.txt
- WangShuo_stewardess.txt
- WangShuo_theTroubleShooter.txt


```python
# define a function to get .txt files in a folder

from os import listdir
def list_textfiles(directory):
    "Return a list of filenames ending in '.txt' in DIRECTORY."
    textfiles = []
    for filename in listdir(directory):
        if filename.endswith(".txt"):
            textfiles.append(directory + "/" + filename)
    return textfiles   

# define a function to read the text in a .txt file

def read_txt(filename):
    try:
        f = open(filename,'r') #open(filename,'r')
        text = f.read()
    finally:
        if f:
            f.close()
    return text

#import harry potter textfiles
filenames = list_textfiles('Fictions')
raw_texts = []
for n in filenames:
    raw_texts.append(read_txt(n))

# how many fictions to be analyzed?
print (len(raw_texts))
```

### Preprocessing
- remove stopwords
- tokenizing

```python
import jieba
# segment text with 'jieba'
# define a function for tokenization

def tokenize (text):
    tokens = jieba.cut(text,cut_all=False)
    # return a list of tokens
    return tokens
```

```python
# import a list of pre-defined stopwords
stopwords = [line.strip() for line in open('stopwords.txt', 'r').readlines()]
# define a function to remove stopwords
def remove_sw(tokens):

    removeST_tokens = []
    for token in tokens:
        if token not in stopwords:
            if token != '\t' and token != '\n':
                removeST_tokens.append(token)
    # return a list of tokens removed stopwords
    return removeST_tokens
    
```

```python
# tokenizing and removeing stopwords
clean_text = []
for text in raw_texts:
    clean_text.append(remove_sw(tokenize(text)))
```

```python
# let's preview the cleaned text
print (clean_text[1])
```

### Vectorization

```python
#vectorize text
text_string = []
for text in clean_text:
    text_string.append(" ".join(text))

from sklearn.feature_extraction.text import CountVectorizer  
n_features = 1000
tf_vectorizer = CountVectorizer(min_df = 2,
                                strip_accents = 'unicode',
                                max_features=n_features,
                                stop_words='english')
tf = tf_vectorizer.fit_transform(text_string)
```

### LDA topic modeling

```python
from sklearn.decomposition import LatentDirichletAllocation
n_topic = 7
lda = LatentDirichletAllocation(n_components = n_topic, 
                                learning_method='online',
                                max_iter=50,
                                random_state=0)
doctopic = lda.fit(tf)

# topic_distribution is a distribution of the topics in each text
topic_distribution = lda.transform(tf)

print (topic_distribution) # not normalized (sum of each row is not 1)

```

```python
def print_top_words(model, feature_names, n_top_words):
    
    #print top terms in each topic
    for topic_idx, topic in enumerate(model.components_):
        print ("Topic #%d:" % topic_idx)
        print (" ".join([feature_names[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]))

# set the top n words in each topic
n_top_words=20
        
tf_feature_names = tf_vectorizer.get_feature_names()
print_top_words(lda, tf_feature_names, n_top_words)
```

- The list above is a probability of topic distrubition in the nine texts.

- Next we will try to visualize the topic distribution in a heatmap.

```python
import matplotlib.ticker as ticker
import matplotlib.cm as cm
import matplotlib as mpl
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

flight_matrix = topic_distribution 

yLabel = list_textfiles('Fictions')
xLabel = ['topic0','topic1','topic2','topic3','topic4','topic5','topic6']

fig = plt.figure()
fig, ax = plt.subplots(1,1, figsize=(6,6))


ax.set_xticks(np.arange(len(xLabel)))
ax.set_yticks(np.arange(len(yLabel)))

ax.set_xticklabels(xLabel)
ax.set_yticklabels(yLabel)

heatplot = ax.imshow(flight_matrix, cmap='Purples')
```
